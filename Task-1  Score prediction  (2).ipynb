### Prediction using supervised ML

#### Simple Linear Regresson implementaion model
In this regression task we predict the percentage of marks that is expected to score based upon the number of hours they studied. This is a simple linear regression task as it involves just two variables.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("data.csv")
df
```

```python
%matplotlib inline
plt.xlabel("Hours")
plt.ylabel("Scores")
plt.scatter(df.Hours,df.Scores,color='red',marker='+')
```

```python
df1=df.drop('Scores',axis='columns')
df1
```

```python
score=df.Scores
score
```

```python
model=linear_model.LinearRegression()
model.fit(df1,score)
```

### Predict the score with hour=9.25

```python
model.predict([[9.25]])
```

```python
model.coef_
```

```python
model.intercept_
```

#### y = m*x+b (m is coefficient and b is intercept)

```python
9.25*9.77580339+2.48367340537321
```

#### Generate the CSV file with list of home price prediction

```python
df_hours=pd.read_csv("data1.csv")
df_hours
```

```python
s=model.predict(df_hours)
s
```

```python
df_hours['Scores']=s
df_hours
```

```python
df_hours.to_csv("prediction.csv")
```
